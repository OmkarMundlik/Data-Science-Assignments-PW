# Q2. Write a python program to print a python exception hierarchy.

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(Exception)
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a base class for exceptions raised when a key or index used to access a container is not found. It is the parent class of several other exception classes, including IndexError, KeyError, and AttributeError.

KeyError and IndexError are both subclasses of LookupError and are used when a key or index is not found in a container, respectively. Here are some examples to illustrate their usage:


```python
# Example of KeyError
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']
except KeyError as e:
    print(f'KeyError: {e}')  # KeyError: 'd'

# Example of IndexError
my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print(f'IndexError: {e}')  # IndexError: list index out of range
```

# Q6. List down some best practices for exception handling in python.
Here are some best practices for exception handling in Python:

Catch specific exceptions: Catch only the exceptions that you expect to be raised and let other exceptions propagate upwards. Catching specific exceptions will help you handle errors more effectively and provide more meaningful error messages.

Keep error messages clear and concise: Error messages should be clear and concise, making it easier for the user to understand what went wrong and how to fix it.

Use try-except-else blocks: Use try-except-else blocks to handle exceptions and perform additional actions if no exceptions are raised. This makes the code more readable and easier to maintain.

Use finally blocks for cleanup: Use finally blocks to perform cleanup actions, such as closing files or releasing resources, regardless of whether an exception is raised or not.

Don't suppress exceptions: Avoid suppressing exceptions by catching them and not doing anything with them. If you catch an exception, you should always handle it in some way, even if it is just logging the error.

Don't catch exceptions that you can't handle: Don't catch exceptions that you can't handle or that you don't know how to handle. This can lead to silent errors and make debugging more difficult.

Use context managers: Use context managers, such as the with statement, to handle resources that need to be cleaned up, such as files or database connections. This makes the code more readable and easier to maintain.

Raise exceptions with clear messages: When raising exceptions, provide clear and concise error messages that explain the reason for the exception and what action the user can take to fix the error.

Use logging: Use the Python logging module to log exceptions and other errors. This makes it easier to track down bugs and diagnose problems in production.

Test your code with various inputs: Test your code with various inputs to ensure that it handles exceptions correctly and provides meaningful error messages. This will help you catch bugs early and avoid problems in production.
