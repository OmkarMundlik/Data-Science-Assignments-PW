# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plots are the graphs that represent the relationship between two variables in a data-set. It presents dta points on a 2D plane. The independent variables  are plotted on X axis where as depenedent variables are plotted on Y axis.


```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
colors = [i for i in range(50)]
plt.scatter(x,y,c = colors)
plt.title("Scatter plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

The pictorial representation of grouped data, in the form of vertical or horizontal rectangular bars, where the lengths of the bars are equivalent to the measure of data, are known as bar graphs or bar charts.

It shows the relationship between a numeric and a categoric variable.
Each entity of the categoric variable is represented as a bar. 
The size of the bar represents its numeric value.
The bar graph is used to compare the items between different groups over time. Bar graphs are used to measure the changes over a period of time. When the changes are larger, a bar graph is the best option to represent the data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(14,6))
plt.subplot(121)
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Vertical Bar Plot")
plt.subplot(122)
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar plot")
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

When we display the data distribution in a standardized way using 5 summary â€“ minimum, Q1 (First Quartile), median, Q3(third Quartile), and maximum, it is called a Box plot.

It is used to know:

The outliers and their values
Symmetry of Data
Tight grouping of data
Data skewness â€“ if, in which direction and how

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig = plt.figure(figsize=(10,7))
plt.boxplot([box1, box2])
plt.show()
```
