# Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask is a micro web framework written in Python that is designed to be lightweight and flexible, with minimal dependencies. It is a popular choice for building small to medium-sized web applications and APIs.

Here are some advantages of Flask:

Simplicity: Flask is simple and easy to use, with a small core and a straightforward API that makes it a great choice for building small web applications or prototypes.

Flexibility: Flask is highly customizable and can be extended with a wide range of third-party libraries and plugins. This allows developers to add additional features to their applications as needed.

Minimalism: Flask has minimal dependencies and does not impose any particular way of doing things. This allows developers to build applications that fit their own preferred style or approach.

Modular: Flask is composed of small, independent components that can be used together or separately, making it a great choice for building modular applications.

Easy to learn: Flask has a small learning curve and is well-documented, making it an ideal choice for beginners who are just starting out with web development.

Testing: Flask has built-in support for testing, making it easy to write and run tests for your application.

Community: Flask has a large and active community of developers who contribute to its development and provide support through forums, mailing lists, and other channels.

Overall, Flask is a great choice for building small to medium-sized web applications and APIs that are lightweight, flexible, and easy to maintain. Its simplicity, flexibility, and modularity make it a popular choice among developers who value these qualities in a web framework.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def helloWorld():
    return "<h1>Hello World!!</h1>"

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

# Q3. What is App routing in Flask? Why do we use app routes?
In Flask, app routing refers to the process of mapping URLs to specific functions or methods in your Flask application. It allows you to define different endpoints or routes for your application, each with its own associated function or method.

App routing is an essential part of building a Flask application because it allows you to define the different URLs that users can visit and the corresponding actions that should be taken when they do so. This allows you to build a logical and intuitive navigation structure for your application and provide a clear and consistent user experience.

Using app routes also makes it easier to separate the logic of your application into smaller, more manageable pieces. By defining different functions or methods for each app route, you can keep your code organized and maintainable, which is especially important for larger applications.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
    In flask, to build a URL for specific function, url_for() function is used. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/")
def helloWorld():
    return "<h1>Hello World!!</h1>"

@app.route("/greet")
def greet():
    return "<h1>Namaste India</h1>"

@app.route("/about")
def about() :
    return "<h1>This is about page</h1>"

@app.route("/user/<name>")
def user(name) :
    if name == "greet" :
        return redirect(url_for("greet"))
    if name=="about" :
        return redirect(url_for("about"))
if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
