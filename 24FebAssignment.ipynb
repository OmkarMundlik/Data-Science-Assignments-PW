# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

# Q5. How are DataFrame.size() and DataFrame.shape() different?
    Both DataFrame.size() and DataFrame.shape() are methods in pandas library used to get information about the dimensions of a DataFrame.

    DataFrame.size() method returns the total number of elements in a DataFrame. In other words, it gives the product of the number of rows and columns in the DataFrame. For example, if a DataFrame has 3 rows and 4 columns, the size of the DataFrame would be 12.

    DataFrame.shape() method returns a tuple containing the number of rows and columns in a DataFrame. The tuple has two elements, where the first element represents the number of rows and the second element represents the number of columns. For example, if a DataFrame has 3 rows and 4 columns, the shape of the DataFrame would be (3, 4).

    In summary, DataFrame.size() returns a single integer representing the total number of elements in the DataFrame, while DataFrame.shape() returns a tuple representing the number of rows and columns in the DataFrame.


```python
df = pd.DataFrame({
    'key1' : [1,2,3,4,5,6],
    'key2' : [2,3,4,5,6,7]
})
print("Size of dataset : ", df.size)
print("Dimensions of the dataset : ", df.shape)
```

# Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains emailaddresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python
df = pd.DataFrame({"Email" : ["omkar.mundlik@gmail.com", "uddhubeta@vit.com", "pritam1234@pritamjwellers.com", "sudhanshu1122@pwskills.edu"]})

def extractUsername(df) :
    df["Username"]=[(i.split("@"))[0] for i in df["Email"]]
    return df

extractUsername(df)
```

# Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.
For example, if df contains the following values:
A B C
0 3 5 1
1 8 2 7
2 6 9 4
3 2 3 5
4 9 1 2

```python
def selectRows(df) :
    df = df[(df['A']>5) & (df['B'] < 10)]
    return df

df = pd.DataFrame({"A" : [3,8,6,2,9], "B": [5,2,9,3,1], "C":[1,7,4,5,2]})


selectRows(df)
```

# Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
df = pd.DataFrame({"Values" : [3,4,5,6,7,8]})

def Statictics(df) : 
    df2 = df.describe()
    mean = df2['Values'][1]
    standardDeviation = df2['Values'][2]
    df.sort_values(by='Values')
    l = len(df['Values'])
    if(l%2==0):
        median = (df['Values'][int(l/2)] + df['Values'][int(l/2) -1])/2
    else:
        median = df['Values'][int(l/2)]
    return [mean, standardDeviation, median]

Statictics(df)
```

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
def movingaverage(x):
    j=0
    sum=0
    for i in df[df['Date'] <= x]['Sales'][::-1] :
        j+=1
        sum += i
        if j==7:
            break
    sum /= 7
    return sum

df = pd.DataFrame({
    'Sales': [100, 200, 150, 300, 250, 175, 225, 275, 125, 350],
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05','2022-01-06', '2022-01-07','2022-01-08', '2022-01-09', '2022-01-10']
})
df['Date'] = pd.to_datetime(df['Date'])

df['MovingAverage'] = df['Date'].apply(movingaverage)
df
```

# Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

```python
dates = pd.date_range('2023-01-01', '2023-01-31', freq='D')
df = pd.DataFrame({'Date' : dates})
def setWeekDay(df):
    df["Weekday"] = df.Date.dt.day_name()
    return df

setWeekDay(df).head()
```

# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.
    

```python
dates = pd.date_range('2022-01-01', '2023-12-31', freq='D')
df = pd.DataFrame({'Date' : dates})

def selectDates(df):
    df2 = df[(df["Date"] >= '2023-01-01') & (df["Date"] <= '2023-01-31')]
    return df2

selectDates(df).head()
```
